# 1. ssh 连接 Linux 服务器

本地远程登录 Linux 服务器，需要使用端口号为22的SSH协议。通常情况下正常安装 Linux 系统之后，SSH 协议是默认放开的，但是也有一些情况需要我们自己安装。

SSH 协议是通过 sshd 服务提供交互的，因此如果想知道 Linux 中有没有 sshd 服务，可以通过尝试启动 sshd 服务。

```shell
chenkc@chenkc:~$ sudo service sshd start
Failed to start sshd.service: Unit sshd.service not found.
chenkc@chenkc:~$
```

如果启动失败说明你的 Linux 服务器中没有安装 sshd 服务，可以通过下面命令进行安装：

```
chenkc@chenkc:~$ sudo apt-get install openssh-server
[sudo] password for chenkc: 
Reading package lists... Done
...
Do you want to continue? [Y/n] # 输入y继续
...
```

使用如下命令查看 sshd 服务是否安装成功：

```
chenkc@chenkc:~$ ps -ef | grep sshd
root       3302      1  0 17:47 ?        00:00:00 /usr/sbin/sshd -D # 安装成功并启动
chenkc     3400   2319  0 17:49 pts/0    00:00:00 grep --color=auto sshd
chenkc@chenkc:~$ 
```

开启和关闭 sshd 服务也非常简单：

```
chenkc@chenkc:~$ service sshd start # 开启sshd服务
chenkc@chenkc:~$ service sshd stop # 停止sshd服务
```

Linux 服务器开启了 sshd 服务，接下来只需要查看 Linux 服务器的 ip 地址，就可以在本地通过 SSH 远程连接 Linux 服务器。我的 Linux 服务器的 ip 地址为：`192.168.135.141`

# 2. 下载 Anaconda 安装包并安装

使用本地机器（因为 Linux 服务器没有图形化界面）访问 Anaconda 官方下载地址：https://www.anaconda.com/products/individual，获取 Linux 版本的 Anconda 的下载地址。这里选择 Python3.8 版本，X86和Power8 and Power9表示不同的处理器架构，这里选择x86即可。

![](https://gitee.com/fendoudaniu/image/raw/master/Deep-Learning-with-PyTorch-Tutorials-master/%E8%8E%B7%E5%8F%96anaconda%E9%93%BE%E6%8E%A5.gif)

使用`wget 下载链接`命令下载 Anaconda 的安装包，其中下载链接就是上面复制的地址：

![](https://gitee.com/fendoudaniu/image/raw/master/Deep-Learning-with-PyTorch-Tutorials-master/Snipaste_2020-10-12_18-56-43.png)

下载完成之后，可以用户家目录中找到安装包：

```
chenkc@chenkc:~$ ll
...
-rw-rw-r--  1 chenkc chenkc 576830621 7月  24 02:25 Anaconda3-2020.07-Linux-x86_64.sh
```

有了 Anaconda 的安装包，接下来安装 Anaconda 就非常简单了，不过由于上传后的文件没有可执行权限，因此需要先为文件添加可执行权限。

![](https://gitee.com/fendoudaniu/image/raw/master/Deep-Learning-with-PyTorch-Tutorials-master/Snipaste_2020-10-12_20-20-39.png)

文件有了可执行权限，直接执行"Anaconda3-2020.07-Linux-x86_64.sh"文件即可进入安装程序。

![](https://gitee.com/fendoudaniu/image/raw/master/Deep-Learning-with-PyTorch-Tutorials-master/Snipaste_2020-10-12_20-22-07.png)

阅读软件许可，直接按"q"即可，然后输入"yes"表示接受许可。

![](https://gitee.com/fendoudaniu/image/raw/master/Deep-Learning-with-PyTorch-Tutorials-master/Snipaste_2020-10-12_20-23-57.png)

指定 Anaconda 的安装路径。

是否添加 Anconda 的环境变量？**为了能够在全局使用，输入"yes"添加即可。**
